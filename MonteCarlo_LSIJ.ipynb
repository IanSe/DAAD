```python
import numpy as pn
import matplotlib.pyplot as mpl
```

```python
def rango():
    x = pn.random.uniform(0,1,1000)
    y = pn.random.uniform(0,1,1000)

    z = []

    for i in range(1000):
        res = (x[i]**2)+(y[i]**2)
        z.append(res)


    contPOS=0
    contNEG=0

    for j in z:
        if j < 1:
            contPOS=contPOS+1
        else:
            contNEG=contNEG+1
            
    prob = contPOS/1000

    return prob
```

```python
aproxPI=[]

for i in range(100):
    aproxPI.append(rango()*4)
    
suma = 0

for num in aproxPI:
    suma = suma + num


avg = suma/len(aproxPI)

print(avg)
```
